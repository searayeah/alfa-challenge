```
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error

df_transactions = pd.read_parquet("data/df_transaction.pa")
df_train = pd.read_parquet("data/train.pa")
```

```
test_clients = list(
    set(df_transactions["client_num"].unique().tolist()).difference(
        df_train["client_num"].to_list()
    )
)
```

```
df_test_baseline = pd.DataFrame(
    {"client_num": test_clients, "target": df_train["target"].mode().to_list()[0]}
)
df_test_baseline
```

```
df_test_baseline.to_csv("submissions/baseline.csv", index=False)
```
