```
import pandas as pd
```

```
tsfresh_features = pd.read_pickle("data/features/tsfresh_generations.pkl")
coles_features = pd.read_pickle("data/features/coles.pkl")
heuristics = pd.read_pickle("data/features/heuristics.pkl")

coles_features = coles_features.reset_index()
coles_features = coles_features.rename(columns={"index": "client_num"})
```

```
full_features = tsfresh_features.merge(coles_features, on="client_num", how="left").merge(
    heuristics, on="client_num", how="left"
)
full_features
```

```
full_features = full_features.fillna(0)
```

```
full_features.to_pickle("data/features/full_features.pkl")
```
